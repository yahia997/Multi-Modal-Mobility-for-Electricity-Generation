```python
import os, glob
import pandas as pd
import geopandas as gpd
```

```python
# Load roads polygons
roads = gpd.read_file('../../data/france/gis_osm_roads_free_1.shp').to_crs(epsg=2154)
```

```python
input_file = "../../data/gps_dataset/*.csv"
output_file = '../../data/gps_road_merged_v2.csv'
header_written = False
```

```python
# Delete the output file if it exists
if os.path.exists(output_file):
  os.remove(output_file)
  print(f"Deleted existing file: {output_file}")
```

```python
all_files = glob.glob(input_file)

all_dfs = []

i = 0
for path in all_files:
  df = pd.read_csv(path)

  # Add ID field (file name)
  id = os.path.basename(path).split(".")[0][-7:]
  df['ID'] = id

  # Join with roads ------------------------------------------------------------
  # Create geometry points and convert to geopandas to be able to perfom the spatial join
  gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df['LONGITUDE'], df['LATITUDE']), 
    crs="EPSG:4326"
  )

  # Spatial join roads
  # map each gps point to single road (nereast road)
  gdf = gdf.to_crs(epsg=2154)
  joined = gpd.sjoin_nearest(
    gdf, 
    roads, 
    how="left", 
    distance_col="distance_to_road"
    )\
  [['LATITUDE', 
    'LONGITUDE', 
    'SPEED', 
    'LOCAL DATETIME', 
    'VALID', 
    'UTC DATETIME', 
    'ID', 
    'fclass', 
    'name', 
    'index_right',
    'distance_to_road'
  ]] 
  # select specific cols for performance
  
  # Write the result directly to the output file to reduce memory usage
  joined.to_csv(output_file, mode='a', header=not header_written, index=False)
  header_written = True  # Ensure header is written only once

  # for logging
  if i%10 == 0:
    print(f'Patch number {i}')
  i += 1
```
